## Experiment Design Notes

These experiments are proposed A/B tests based on historical behavioural analysis.  
The dataset does not include live experiment assignment data, so the notebook does not claim actual experiment results.

Each test includes:

- A target segment
- A clear hypothesis
- A control and variant
- A primary success metric
- Secondary and guardrail metrics
- A decision rule for rollout
